```python
import numpy as np
```

```python
#array
x = np.array([[1, 3, 7],
              [4, 11, 21],
              [42, 8, 9]])
print("x:\n", x)
```

```python
#reshape: transformar a matriz em um vetor coluna
#(3, 3) vira (9,1): 3*3 = 9*1 =9
print("transformação de uma matriz em um vetor coluna:\n", x.reshape(9, 1))
```

```python
#reshape: transformar a matriz em um vetor linha
#(3, 3) vira (1, 9): 3*3 = 9*1 =9
print("transformação de uma matriz em um vetor linha:\n", x.reshape(1, 9))
```

```python
#transposição de matriz
print("x transporta:\n", x.T)
```

```python
#np.sum: soma em um dado eixo, axis = {0:linha, 1: coluna}
print("x:\n", x)
print("soma de todos elementos de x:", np.sum(x))
print("soma x ao logo das  linhas:", np.sum(x, axis=0))
print("soma x ao logo das colunas:", np.sum(x, axis=1))

```

```python
#np.mean: media em um dado eixo, axis = {0:linha, 1: coluna}
print("x:\n", x)
print("média de todos elementos de x:", np.mean(x))
print("média x ao logo das  linhas:", np.mean(x, axis=0))
print("média x ao logo das colunas:", np.mean(x, axis=1))

```

```python
#np.where, indentificação dos indíces onde uma dada condição (coordenadas)
# é atendida. Uso com indexação boolena
cond = x % 2 == 0 #números pares
print("condição:\n", cond)
i, j = np.where (cond) #indices x[i, j]= x[cond]
print("índice i (linhas)", i)
print("índice j (colunas)", j)

```

```python
#indexação booleana e slicing: selecionar as linhas
# de x que possuem algum número par
print("x:\n", x)
cond =  x %2 == 0
print("condição:\n", cond)
#se houver alguma condição True na linha,  a soma será>0
i_row = np.where(np.sum(cond, axis=1))[0]
print ("indices das linhas que possuem números pares:\n", i_row)
print ("linhas que possuem números pares:\n", x[i_row, :])
```

```python
#regressão linear
#Determina o comportamento de duas variáveis;
# em um problema linear: y = f(x) = ax + b ;   a e b são  coeficentes.

```

```python
#visualização de dados
import matplotlib.pyplot as plt
```

```python
#dados
x = [-1., -0.7777778, -0.55555556,-0.333333333, -0.1111111, 0.11111111, 0.33333333, 0.55555556, 0.77777778, 1]
y = [-1.13956201, -0.57177999, -0.21697033, 0.5425699, 0.49406657,1.14972239, 1.64228553, 1.64228553, 2.1749824, 2.95684202]
```

```python
#plot dos dados
plt.figure(figsize=(10,5))
plt.plot (x, y,'o', label='dados originais')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
"""
iremos estimar uma função do tipo : y =a*x + b
os valores reis de (a,b): 2 , 1
"""
#transformando para numpy e vetorcoluna
x, y = np.array(x).reshape(-1,1), np.array(y).reshape(-1,1)

# adicionando bias: para estimar o termo b
x = np.hstack((x, np.ones(x.shape)))

#estimando a e b
beta = np.linalg.pinv(x).dot(y)
print("a estimado:", beta[0][0])
print("b estimado:", beta[1][0])

```

```python
#plot dos dados
plt.figure(figsize=(10,5))
plt.plot (x, y,'o', label='dados originais')
plt.plot(x, x.dot(beta), label='regressão linear')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title ("Regressão linear com numpy")
plt.grid()
plt.show()
```
